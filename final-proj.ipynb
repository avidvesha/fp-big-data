```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
from urllib.parse import quote
import re
import time
import requests
```

```python
# Create URLs for all pages
pages = []
for i in range(0, 59):
    url = f"https://www.tc-v.com/used_car/audi/all?pn={i}"
    pages.append(str(url))

pages
```

```python
# Scrape all food links from pages
links = []
for page in pages:
    try:
        response = requests.get(page)
        soup = BeautifulSoup(response.text, 'html.parser')
        for link in soup.find_all("div", {"class": "car-item__pic-area"}):
            for a in link:
                links.append(a.get('href'))
    except Exception as e:
        print(f"Error scraping page {page}: {e}")

print(f"Found {len(links)} product links")
links
```

```python
# Scrape details for each food item with error handling
data_cars = []

# Use the second for-loop for testing purpose
for idx, link in enumerate(links):
    try:
        url = f"https://www.tc-v.com{link}"
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Check if elements exist before accessing them
        car_name = soup.find("h1", {"class": "car__detail-ttl"})
        vin = soup.find("div", {"class": "vin__main-text"})
        details = soup.find_all("td", {"class": "car__info-table-body"})
        
        data_car = {
            'name': car_name.text.strip(),
            'vin': vin.text,
            'model_code': details[1].text,
            'registration_year': details[2].text,
            'manufacture_year': details[3].text,
            'mileage': details[4].text,
            'transmission': details[5].text,
            'engine_capacity': details[6].text,
            'fuel': details[7].text,
            'body_style1': details[8].text,
            'body_style2': details[9].text,
            'steering': details[10].text,
            'ext_color': details[11].text,
            'int_color': details[12].text,
            'drive_type': details[13].text,
            'door': details[14].text,
            'seat': details[15].text,
            'dimension': details[16].text,
            'condition': details[17].text,
            'id': details[18].text,
            'remarks': details[19].text,
            'comment': details[20].text,
            'expiry_date': details[21].text
        }
        data_cars.append(data_car)
        print(f"Successfully scraped product {idx+1}/{len(links)}: {data_car['name']}")

        # Add a small delay to avoid overloading the server
        time.sleep(0.5)

    except Exception as e:
        print(f"Error scraping link {link}: {e}")

print(f"Successfully scraped {len(data_cars)} products out of {len(links)} links")
```

```python
# Convert the list of dictionaries to a DataFrame
df_cars = pd.DataFrame(data_cars)
print(f"Jumlah produk berhasil di-scrape: {len(df_cars)}")
df_cars
```

```python
# Export to CSV
df_cars.to_csv('audi_cars_data.csv', index=False)
print("Data exported to audi_cars_data.csv")

```
